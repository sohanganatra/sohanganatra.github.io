###Scraping The Data

To begin, we had get the data from the website into a usable format. Since we did not have a formatted file like a csv ready to read the data from, we had to start by scraping the data from the website. 

The first thing we noticed was that the data was not held in a table within the html. It was all in a long text string. Irregularities in the data brought up many challenges in parsing the string. Some columns had data that was more than one word, so we could not simply break the string into words and fill the dataframe from there. Additionally, some columns were blank, which we aso had to account for that when parsing the string into data.

The following are methods that helped parse the string into data and create a dataframe.

```python
def identifier(a): # takes in an array of strings and returns it as a list 
                   # with sublists containing the values a row
    rowends = []
    rowstarts = []
    rowstarts.append(0)
    count = 0
    i = 0
    # Uses the date as a point for starting new lines, because each row contains a formatted date
    while i < len(a):
        if re.search("\d\d\.\d\d\.\d\d\d\d", a[i]):
            rowends.append(i)
            if (i + 1) < len(a):
                rowstarts.append(i + 1)
        i += 1
    rows = []
    
    for i in range(0, len(rowends)):
        rowi = [] #initial row list
        rown = [] #rows with names corrected
        rowb = [] #rows with dates corrected assuming names are right
        rowl = [] #rows with location corrected assuming names and birthdays are right
        
        # make rowi
        for j in range(rowstarts[i], rowends[i] + 1):
            rowi.append(a[j])
            j += 1
        
        # combine names
        j = 0
        while j < len(rowi):
            if j == 2:
                rown.append(rowi[j])
                j += 1
                while not (re.search("[A-Z][A-Z][A-Z]", rowi[j])):
                    rown[2] = rown[2] + " " + rowi[j]
                    j += 1
            else:
                rown.append(rowi[j])
                j += 1
        
        # fix dates
        j = 0
        while j < len(rown):
            if j == 4:
                if re.search("\d\d\.\d\d\.\d\d", rown[j]):
                    rowb.append(rown[j])
                    j += 1
                else:
                    if re.search("[A-Z][a-z]*", rown[j + 1]):
                        rowb.append("-")
                        rowb.append(rown[j])
                        j += 1
                    else:
                        rowb.append("-")
                        j += 1
                        
            else:
                rowb.append(rown[j])
                j += 1
        
        # combine locations
        j = 0
        while j < len(rowb):
            if j == 6:
                rowl.append(rowb[j])
                j += 1
                while j < len(rowb) and not (re.search("\d\d\.\d\d\.\d\d\d\d", rowb[j])):
                    rowl[6] = rowl[6] + " " + rowb[j]
                    j += 1
            else:
                rowl.append(rowb[j])
                j += 1
        
        rows.append(rowl)
            
    return rows


def df_event(soup): # Takes in the BeautifulSoup object and returns a datafram
    # All of the data we want is in a <pre> tag, so we find that directly
    string = soup.find('pre')
    string = string.get_text()
    rows = string.split() # Split text on whitespace
    rows = identifier(rows) # Call above identifier method which will format string into list of rows
    df = pd.DataFrame(columns=["Rank", "Time", "Name", "Country", "Date of Birth", "Place", "City", "Date"]) 
    #for i in range(0, len(rows_10)):
    #    if len(rows_10[i]) != 8:
    #        print(rows_10[i])
    for i in rows:
        df.loc[len(df.index)] = i
    return df
```

###Cleaning The Data



Now that we have the off of the website placed into dataframes, we can tidy up some parts of it so that the data is easier to work with.

First off, we want to convert the times for the different distances into the same format, so they can be compared more easily. We added a column to each dataframe that would hold the time in seconds. 

The methods below take in the time in various formats and return the time in seconds.

```python
# Converts time from MM:SS.SS to seconds for 5000m and 10000m
def secs_510(time):
    if re.search("(\d\d):(\d\d\.\d\d?)", time):
        result = re.search("(\d\d):(\d\d\.\d\d?)", time)
        return (float(result.group(1)) * 60.0) + float(result.group(2))

# Converts time from MM:SS to seconds for half marathon
def secs_hm(time):
    if re.search("(\d\d):(\d\d)", time):
        result = re.search("(\d\d):(\d\d)", time)
        return (float(result.group(1)) * 60.0) + float(result.group(2))

# Converts time from M:SS.SS to seconds for 1500m and 3000m
def secs_153(time):
    if re.search("(\d):(\d\d\.\d\d?)", time):
        result = re.search("(\d):(\d\d\.\d\d?)", time)
        return (float(result.group(1)) * 60.0) + float(result.group(2))

# Converts time from H:MM:SS to seconds for marathon
def secs_m(time):
    if re.search("(\d):(\d\d):(\d\d)", time):
        result = re.search("(\d):(\d\d):(\d\d)", time)
        return (float(result.group(1)) * 60.0 * 60.0) + (float(result.group(2)) * 60.0) + float(result.group(3))

# Adding the time in seconds to a new column in each dataframe
df_10['Time_Sec'] = df_10["Time"].map(lambda element : secs_510(element))
df_15['Time_Sec'] = df_15["Time"].map(lambda element : secs_153(element))
df_hm['Time_Sec'] = df_hm["Time"].map(lambda element : secs_hm(element))
df_5['Time_Sec'] = df_5["Time"].map(lambda element : secs_510(element))
df_m['Time_Sec'] = df_m["Time"].map(lambda element : secs_m(element))
df_3['Time_Sec'] = df_3["Time"].map(lambda element : secs_153(element))
```

Next, we wanted to make an alternate dataframe for each dataset that would hold the top performers rather then all of the top performances. As mentioned above, the original dataset conatains all of the top performaces, including multiple performaces by the same athlete. In most cases, we will want to only look at each athlete's best performance, since we will be looking at how an athlete's peak ability in shorter events can predict their time in the marathon. Our data also is limited and we can not guaruntee to have the same number of performances from each athlete. Some athletes have a lot of performances on the top table and some have only one. Having different numbers of permformance from the different athletes could skew out data.

So we made a PB table that keep only the top time for each athlete and removes duplicates on name. We also added a new column with rank which contains the new rank in the top performers table.

```python
def rank_pbs(df): # Method adds the new rank to the table
    df = df.drop(columns = ['Rank']).reset_index().drop(columns = ['index']).reset_index()
    df['Rank'] = df['index'].apply(lambda elem : elem + 1)
    df = df.drop(columns = ['index'])
    return df

# Drops duplicates entries on name and passes into the rank_pbs method to add the new rank
df_10_pb = rank_pbs(df_10.drop_duplicates(subset = ['Name']))
df_15_pb = rank_pbs(df_15.drop_duplicates(subset = ['Name']))
df_hm_pb = rank_pbs(df_hm.drop_duplicates(subset = ['Name']))
df_m_pb = rank_pbs(df_m.drop_duplicates(subset = ['Name']))
df_5_pb = rank_pbs(df_5.drop_duplicates(subset = ['Name']))
df_3_pb = rank_pbs(df_3.drop_duplicates(subset = ['Name']))

df_m_pb.head()
```

### **Consistancy Metric**

While rank gave us a more standardized comparison between the events, we wanted to be able to have a standard way to compare the performances in the differnet events based on time. We decided to create a standardized metric that would be based on times instead of place. 
Another consideration we thought of was consistancy. The top performer's rankings are based on the athlete's personal best time. But we also wanted to factor in the number of top performances an athlete had posted. For example, if someone has run consistantly run fast times, that could indicate better overall performance than someone who has run an exceptionally fast time just once but could not recreate it. That said, we still want to value faster performances higher than slower ones. Someone who has five performances in the top 50 indicates better performance than someone who has run 5 performances in the bottom of the table.

To achieve this we created a metric that we call consistancy score. First we split split up the times into five bins by splitting the range of times in the data into 5 even intervals. For every performance an athlete had in the fastest bin, they get 5 points. For every performance in the second fasteset bin they get 4 points, and so on down to 1 point. This way an athlete gets rated higher based on both faster performances and repeating those fast performances.

The way the score was calculated was by taking the performers from the top perfomers (PB) table, checking their performances in the all top performances table. 

For the consistency score, we also decided to limit the data. Since all the datasets were of different size, having performers for certain distances would skew the points. Based on the smallest dataset, we decided to go with only the top 500 performers.

The following methods add up the score this way for each athlete in the table for each event. 

```python
def total_points_5(name): # Takes an athlete name as input and calculates their consistency score for the 5000m
    def points_5(time):

        # The ranges were identified by cutting dataframe in a seperate step
        if 755.32 < time <= 763.354:
            return 5
        elif 763.354 < time <= 771.348:
            return 4
        elif 771.348 < time <= 779.342:
            return 3
        elif 779.342 < time <= 787.336:
            return 2
        elif 787.336 < time <= 795.33:
            return 1
        else:
            return 0
    
    df_5_person = df_5[df_5['Name'] == name]
    points = 0
    for i in df_5_person['Time_Sec'].values:
        points = points + points_5(i)
    return points

def total_points_10(name): # Takes an athlete name as input and calculates their consistency score for the 10000m
    def points_10(time):
        if 1570.909 < time <= 1589.234:
            return 5
        elif 1589.234 < time <= 1607.468:
            return 4
        elif 1607.468 < time <= 1625.702:
            return 3
        elif 1625.702 < time <= 1643.936:
            return 2
        elif 1643.936 < time <= 1662.17:
            return 1
        else:
            return 0
    
    df_10_person = df_10[df_10['Name'] == name]
    points = 0
    for i in df_10_person['Time_Sec'].values:
        points = points + points_10(i)
    return points

def total_points_hm(name): # Takes an athlete name as input and calculates their consistency score for the Half Marathon
    def points_hm(time):
        if 3450.817 < time <= 3487.6:
            return 5
        elif 3487.6 < time <= 3524.2:
            return 4
        elif 3524.2 < time <= 3560.8:
            return 3
        elif 3560.8 < time <= 3597.4:
            return 2
        elif 3597.4 < time <= 3634.0:
            return 1
        else:
            return 0
    
    df_hm_person = df_hm[df_hm['Name'] == name]
    points = 0
    for i in df_hm_person['Time_Sec'].values:
        points = points + points_hm(i)
    return points
```

```python
# Limiting data to the top 500 performers.
df_hm_500 = df_hm_pb[df_hm_pb['Rank'] <= 500]
df_10_500 = df_10_pb[df_10_pb['Rank'] <= 500]
df_5_500 = df_5_pb[df_5_pb['Rank'] <= 500]
df_5_500['Interval'] = pd.cut(df_5_500['Time_Sec'], 5)
df_10_500['Interval'] = pd.cut(df_10_500['Time_Sec'], 5)
df_hm_500['Interval'] = pd.cut(df_hm_500['Time_Sec'], 5)

#Finding the intervals to make the functions. These are commented out because they
# were only used once to figure out the intervals
#print(df_5_500['Interval'].drop_duplicates())
#print(df_10_500['Interval'].drop_duplicates())
#print(df_hm_500['Interval'].drop_duplicates())

# Creates a new column for consistency points in the dataframes
df_5_500['Consistency Points'] = df_5_500['Name'].map(lambda elem : total_points_5(elem))
df_10_500['Consistency Points'] = df_10_500['Name'].map(lambda elem : total_points_10(elem))
df_hm_500['Consistency Points'] = df_hm_500['Name'].map(lambda elem : total_points_hm(elem))
print(df_hm_500.head())


# Totalling Consistency Scores
df_all500vm = pd.merge(df_hm_500.drop(columns = ['Place', 'City', 'Date', 'Rank']), 
                       df_m_pb.drop(columns = ['Place', 'City', 'Date', 'Rank']), how='right', on=['Name', 'Date of Birth', 'Country'])
df_all500vm = pd.merge(df_5_500.drop(columns = ['Place', 'City', 'Date', 'Rank']), df_all500vm, how='right', on=['Name', 'Date of Birth', 'Country'])
df_all500vm = pd.merge(df_10_500.drop(columns = ['Place', 'City', 'Date', 'Rank']), df_all500vm, how='right', on=['Name', 'Date of Birth', 'Country'])
df_all500vm["Consistency Points_x"] = df_all500vm["Consistency Points_x"].fillna(0)
df_all500vm["Consistency Points_y"] = df_all500vm["Consistency Points_y"].fillna(0)
df_all500vm["Consistency Points"] = df_all500vm["Consistency Points"].fillna(0)
df_all500vm['Total Consistency'] = df_all500vm['Consistency Points'] + df_all500vm['Consistency Points_x'] + df_all500vm['Consistency Points_y']
df_all500vm = df_all500vm[df_all500vm['Total Consistency'] != 0]
```

# **Part 3: Can we use this data to predict marathon performances?**

Given this analysis, we wanted to see what metrics would be the most effective indicators of success in the marathon, as well as see if we could create a model that would reliably predict marathon success in athletes.

###Fitting a Linear Regression

We decided to start with a linear regression model on the data comparing the shorter distance times to the marathon times. From the plots above, there seems to be the strongest correlation when directly comparing the times.

We fit the regression to the data, and then used it to predict each athlete's marathon time based on their 5000m, 10000m, and Half Marathon time. For each we made a new datafram that contains a column of the predictions. We then added a column that calculated the residuals. The residuals are how much the model's predicted marathon time was off of the athlete's actual marathon time.


```python
# Creating the x and y to pass into the model
x_5 = df_5vm['5000m Time (sec)'].values.reshape((-1,1))
y_5 = df_5vm['Marathon Time (sec)'].values

x_10 = df_10vm['10000m Time (sec)'].values.reshape((-1,1))
y_10 = df_10vm['Marathon Time (sec)'].values

x_hm = df_hmvm['Half Marathon Time (sec)'].values.reshape((-1,1))
y_hm = df_hmvm['Marathon Time (sec)'].values

# Fitting the model
linear_regr_5 = LinearRegression().fit(x_5, y_5)
linear_regr_10 = LinearRegression().fit(x_10, y_10)
linear_regr_hm = LinearRegression().fit(x_hm, y_hm)

# Creating a new dataframe, and adding to it a column that contains the model's prediction for each person's marathon time based on their 5000m time
df_5vm_pred = df_5vm
df_5vm_pred["Predicted Marathon Time"] = linear_regr_5.predict(x_5)
df_5vm_pred["residuals"] = df_5vm_pred["Marathon Time (sec)"] - df_5vm_pred["Predicted Marathon Time"]

# Creating a new dataframe, and adding to it a column that contains the model's prediction for each person's marathon time based on their 10000m time
df_10vm_pred = df_10vm
df_10vm_pred["Predicted Marathon Time"] = linear_regr_10.predict(x_10)
df_10vm_pred["residuals"] = df_10vm_pred["Marathon Time (sec)"] - df_10vm_pred["Predicted Marathon Time"]

# Creating a new dataframe, and adding to it a column that contains the model's prediction for each person's marathon time based on their Half Marathon time
df_hmvm_pred = df_hmvm
df_hmvm_pred["Predicted Marathon Time"] = linear_regr_hm.predict(x_hm)
df_hmvm_pred["residuals"] = df_hmvm_pred["Marathon Time (sec)"] - df_hmvm_pred["Predicted Marathon Time"]
```

To see how the model performed, we wanted to plit the residuals. We decided to plot the residuals against the year to see how the model's performance changed over the years. We based this off the year that the training time was run.

To do this, we first had to edit the dataframe to contain the year, instead of just the date.

```python
# Getting the year out of the string containing date
df_5vm_pred2 = df_5vm_pred
years_5 = df_5vm_pred2["Date_x"]
# Converting each date string to a datetime object
for i in years_5:
  i = datetime.strptime(i, '%d.%m.%Y')
df_5vm_pred2["Year"] = pd.to_datetime(years_5)
# Extracting year from the datetime object
df_5vm_pred2["Year"] = df_5vm_pred2["Year"].map(lambda e : e.year)

# Getting the year out of the string containing date
df_10vm_pred2 = df_10vm_pred
years_10 = df_10vm_pred2["Date_x"]
# Converting each date string to a datetime object
for i in years_10:
  i = datetime.strptime(i, '%d.%m.%Y')
df_10vm_pred2["Year"] = pd.to_datetime(years_10)
# Extracting year from the datetime object
df_10vm_pred2["Year"] = df_10vm_pred2["Year"].map(lambda e : e.year)

# Getting the year out of the string containing date
df_hmvm_pred2 = df_hmvm_pred
years_hm = df_hmvm_pred2["Date_x"]
# Converting each date string to a datetime object
for i in years_hm:
  i = datetime.strptime(i, '%d.%m.%Y')
df_hmvm_pred2["Year"] = pd.to_datetime(years_hm)
# Converting each date string to a datetime object
df_hmvm_pred2["Year"] = df_hmvm_pred2["Year"].map(lambda e : e.year)
```

Next, we grouped each dataframe by the year. From this, we get a datafram for each event that contains the average residuals of prediction for each year in the dataset. We use this to plot and compare the performance of each model over time.

```python
# Groupby year
df_5_by_year = df_5vm_pred2.groupby(by="Year").mean()
df_10_by_year = df_10vm_pred2.groupby(by="Year").mean()
df_hm_by_year = df_hmvm_pred2.groupby(by="Year").mean()

# Making a lineplot of each distance
sns.lineplot(data=df_5_by_year, x='Year', y='residuals', color='blue')
sns.lineplot(data=df_10_by_year, x='Year', y='residuals', color='magenta')
sns.lineplot(data=df_hm_by_year, x='Year', y='residuals', color='orange')
plt.legend(labels=["5000m Residuals","10000m Residuals","Half Marathon Residuals"])
```
